```python
import scipy.io as sio
import dotenv
import os

import numpy as np

dotenv.load_dotenv()

INPUT_FOLDER = os.getenv("INPUT_FOLDER")


# Read the data from the .npz file
file_path = f"{INPUT_FOLDER}/Toroidal_topology_grid_cell_data/rat_s_grid_modules_1.npz"
data = np.load(file_path, allow_pickle=True)

# Show what is in the loaded file
print("Keys in the npz file:", list(data.keys()))
for key in data:
    print(f"\nKey: {key}")
    arr = data[key]
    print("Shape:", arr.shape)
    # Only try to print the first five entries if array is at least 1-dimensional
    if arr.ndim == 0:
        print("First entry:", arr.item())
    else:
        print("First few entries:", arr[:5])
```
